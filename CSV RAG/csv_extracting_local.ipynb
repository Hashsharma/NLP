```python
from llama_cpp import Llama

llm = Llama(
    model_path="D:/mr_document/all_models/PrunaAI/mistralai-Mistral-7B-v0.1-GGUF-smashed/Mistral-7B-v0.1.Q4_K_S.gguf",
    n_gpu_layers=-1,    # -1 puts all layers on GPU (recommended for RTX 4060)
    n_ctx=2048,         # adjust context size as needed
    verbose=True        # shows detailed loading info
)
```

```python

response = llm("Q: What is the capital of France?\nA:", max_tokens=32)
print(response["choices"][0]["text"])
```

```python

# Define messages in chat format
messages = [
    {
        "role": "system",
        "content": (
            "You are an expert question generator. Your job is to create high-quality, non-repeating multiple-choice questions (MCQs) "
            "for educational purposes. Given a topic, generate original questions that do NOT conflict with previous ones. Each question must include:\n\n"
            "1. A well-formed multiple-choice question related to the topic.\n"
            "2. Four options labeled A, B, C, and D.\n"
            "3. The correct answer (e.g., A, B, C, or D).\n"
            "4. A concise explanation of why the selected option is correct.\n\n"
            "Make sure every question is unique and not a variation of a previously generated question. Do not include any duplicate or similar concepts.\n\n"
            "Generate as many relevant MCQs as you can from the topic provided."
        )
    },
    {
        "role": "user",
        "content": "Topic: Photosynthesis"
    }
]

# Generate the response
response = llm.create_chat_completion(
    messages=messages,
    max_tokens=512,
    temperature=0.7
)

# Output the generated text
print(response["choices"][0]["message"]["content"])
```
