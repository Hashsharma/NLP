```python
import pandas as pd
import ast  # Used to safely evaluate strings as Python objects

# Read the CSV
df = pd.read_csv('../resource/qnadata/qna_unfiltered.csv')

# Convert string to actual list using ast.literal_eval
df['options'] = df['options'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else [])

# Expand each list into its own columns
options_split = df['options'].apply(pd.Series)

# Rename columns as option_A, option_B, etc.
options_split.columns = [f'option_{chr(65 + i)}' for i in range(options_split.shape[1])]

# Combine back with the original DataFrame
df = pd.concat([df.drop(columns=['options']), options_split], axis=1)

# Save to new CSV
df.to_csv('qna_split_options.csv', index=False)

# Show preview
print(df.head())
```
