```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis")
```

```python
clss = classifier("I've been waiting for a HuggingFace course my whole life.")
clss2 = classifier("Its very hard to judge it but movie was bad")
print(clss, clss2)
```

```python
li_cls = classifier([
    "It m feeling very excited today for this NLP tasks", "The place where i went was average"
])
print(li_cls)
```
