```python
from transformers import pipeline
from pprint import pprint
import os
from dotenv import load_dotenv
load_dotenv()
MODEL_PATH = os.getenv("ALL_MODELS_PATH")
```

```python
# This pipeline is called zero-shot because you donâ€™t need to fine-tune the model on your data to use it.
# It can directly return probability scores for any list of labels you want!
FB_PATH = str(MODEL_PATH) + "/classifier_model/bart-large-mnli/"
classifier = pipeline("zero-shot-classification", FB_PATH, local_files_only=True)
```

```python
classify = classifier("This is the information about the Transformers library",
                      candidate_labels=["education", "politics", "business"])
pprint(classify)
```
