```python
# Text generation
from transformers import pipeline
from pprint import pprint
import os
from dotenv import load_dotenv
load_dotenv()
MODEL_PATH = os.getenv("ALL_MODELS_PATH")
```

```python
SMLM2_MODEL = str(MODEL_PATH) + "/SmolLM2-360M/"
generator = pipeline("text-generation", SMLM2_MODEL)
```

```python
generated = generator("Is it easy to learn NLP, or its beginner friendly",
                    max_length=30,
                    num_return_sequences=2)

pprint(generated)
```
